```python
# final GUI

#home GUI
import tkinter as tk
import plotly
import plotly.graph_objects as go
import pandas as pd
from tkinter import *
from PIL import Image, ImageTk
import os
from plotly.subplots import make_subplots
%store -r checking_temp
%store -r checking_hum
%store -r myhome

HEIGHT=800
WIDTH=700

root= tk.Tk()

os.system("welcome.mp3")

def house():
    global my_img, new_pic, new_pic1,new_pic2,new_pic4,new_pic5,new_pic6,new_pic8,new_pic9,new_pic10,new_pic11,new_pic12,new_pic13
    top= Toplevel()
    top.title('House overall status')
    my_img= ImageTk.PhotoImage(Image.open("houseview.png"))
    my_label= Label(top, image=my_img).pack()
    btn= Button(top, text= "Back to dashboard", command=top.destroy).pack()
    #all on
    if myhome==1: # both ON
        my_img2= Image.open("watering.png")
        resized= my_img2.resize((50,50),Image.ANTIALIAS)
        new_pic=ImageTk.PhotoImage(resized)
        #2nd image
        my_img3= Image.open("watering.png")
        resized1= my_img3.resize((50,50),Image.ANTIALIAS)
        new_pic1=ImageTk.PhotoImage(resized1)
        #heater
        my_heat= Image.open("heateron.png")
        resized2= my_heat.resize((110,160),Image.ANTIALIAS)
        new_pic2=ImageTk.PhotoImage(resized2)
        #labelling
        my_label2= Label(top, image=new_pic)
        my_label2.place(relx=0.2, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label3= Label(top, image=new_pic1)
        my_label3.place(relx=0.65, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label4= Label(top, image=new_pic2)
        my_label4.place(relx=0.3, rely=0.45, relwidth=0.03, relheight=0.09)
        os.system("bothON.mp3")
    elif myhome==2: ## heater off, water on
        my_img8= Image.open("watering.png")
        resized8= my_img8.resize((50,50),Image.ANTIALIAS)
        new_pic8=ImageTk.PhotoImage(resized8)
        #2nd image
        my_img9= Image.open("watering.png")
        resized9= my_img9.resize((50,50),Image.ANTIALIAS)
        new_pic9=ImageTk.PhotoImage(resized9)
        #heater
        my_heat10= Image.open("heateroff.png")
        resized10= my_heat10.resize((110,160),Image.ANTIALIAS)
        new_pic10=ImageTk.PhotoImage(resized10)
        #labelling
        my_label8= Label(top, image=new_pic8)
        my_label8.place(relx=0.2, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label9= Label(top, image=new_pic9)
        my_label9.place(relx=0.65, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label10= Label(top, image=new_pic10)
        my_label10.place(relx=0.3, rely=0.45, relwidth=0.03, relheight=0.09)
        os.system("HeaterOffwaterON.mp3")
    elif myhome==3: ## heater on, water off
        my_img11= Image.open("pngplant.png")
        resized11= my_img11.resize((100,100),Image.ANTIALIAS)
        new_pic11=ImageTk.PhotoImage(resized11)
        #2nd image
        my_img12= Image.open("pngplant.png")
        resized12= my_img12.resize((100,100),Image.ANTIALIAS)
        new_pic12=ImageTk.PhotoImage(resized12)
        #heater
        my_heat13= Image.open("heateron.png")
        resized13= my_heat13.resize((110,160),Image.ANTIALIAS)
        new_pic13=ImageTk.PhotoImage(resized13)
        #labelling
        my_label11= Label(top, image=new_pic11)
        my_label11.place(relx=0.2, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label12= Label(top, image=new_pic12)
        my_label12.place(relx=0.65, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label13= Label(top, image=new_pic13)
        my_label13.place(relx=0.3, rely=0.45, relwidth=0.03, relheight=0.09)
        os.system("HeaterONwaterOff.mp3")      
    else: #all off
        my_img4= Image.open("pngplant.png")
        resized4= my_img4.resize((100,100),Image.ANTIALIAS)
        new_pic4=ImageTk.PhotoImage(resized4)
        #2nd image
        my_img5= Image.open("pngplant.png")
        resized5= my_img5.resize((100,100),Image.ANTIALIAS)
        new_pic5=ImageTk.PhotoImage(resized5)
        #heater
        my_heat1= Image.open("heateroff.png")
        resizedh= my_heat1.resize((110,160),Image.ANTIALIAS)
        new_pic6=ImageTk.PhotoImage(resizedh)
        #labelling
        my_label5= Label(top, image=new_pic4)
        my_label5.place(relx=0.2, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label6= Label(top, image=new_pic5)
        my_label6.place(relx=0.65, rely=0.2, relwidth=0.06, relheight=0.06)
        my_label7= Label(top, image=new_pic6)
        my_label7.place(relx=0.3, rely=0.45, relwidth=0.03, relheight=0.09)
        os.system("bothOFF.mp3")   
        
 
   
def LSTM_temp():  
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    excel_file_t = 'forecastLSTM.xlsx'
    df3 = pd.read_excel(excel_file_t)
    excel_file_h = 'forecastLSTM_h.xlsx'
    df4 = pd.read_excel(excel_file_h)
    
    print(df3)
    print(df4)
    
    fig.add_trace(
    go.Scatter(x=df3['Date'],y=df3['Temperature'], name = "LSTM temperature"),
    secondary_y=False,
    )
    fig.add_trace(
    go.Scatter(x=df4['Date'],y=df4['Humidity'], name = "LSTM humidity"),
    secondary_y=False,
    )
    fig.update_layout(
    title_text="Forecast of Temperature and Humidity "
    )
    fig.update_xaxes(title_text="<b>Date</b>")
    fig.update_yaxes(title_text="<b>Values</b>", secondary_y=False)
    #fig.show()
    plotly.offline.plot(fig, filename="LSTM_Output.html")

    
    
def deviation():  
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    excel_file = 'forecastLSTM.xlsx'
    df1 = pd.read_excel(excel_file)
    excel_file2 = 'dev.xlsx'
    df2 = pd.read_excel(excel_file2)
    
    print(df1)
    print(df2)
    
    fig.add_trace(
    go.Scatter(x=df1['Date'],y=df1['Temperature'], name = "LSTM temperature"),
    secondary_y=False,
    )
    fig.add_trace(
    go.Scatter(x=df2['Date'],y=df2['Temperature'], name = "temperature of Pi"),
    secondary_y=False,
    )
    fig.update_layout(
    title_text="Deviation "
    )
    fig.update_xaxes(title_text="<b>Date</b>")
    fig.update_yaxes(title_text="<b>Temperature</b>", secondary_y=False)
    #fig.show()
    plotly.offline.plot(fig, filename="deviation_temperature.html")


    
def raspi():  
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    excel_file1 = 'rasp_input.csv'
    df1 = pd.read_csv(excel_file1)
    print(df1)

    fig.add_trace(
    go.Scatter(x=df1['Date'],y=df1['Temperature'], name = "temperature"),
    secondary_y=False,
    )

    fig.add_trace(
    go.Scatter(x=df1['Date'],y=df1['Humidity'], name="Humidity")
    )

    fig.update_layout(
    title_text="RaspberryPi sensor values "
    )
    fig.update_xaxes(title_text="<b>Date</b>")
    fig.update_yaxes(title_text="<b>values</b>", secondary_y=False)
    #fig.show()
    plotly.offline.plot(fig, filename="Raspberrypi_temperature.html")




canvas1 = tk.Canvas(root, width = WIDTH, height = HEIGHT)
canvas1.place(x=1000,y=1000)
canvas1.pack()

image5 = Image.open("transparent-smart-house-icon-dashboard-icon-home-icon-5d776b2072a478.4403257515681072964696.png")
image5 = image5.resize((150,150), Image.ANTIALIAS)
photo5 = ImageTk.PhotoImage(image5)

image1 = Image.open("neuralnetwork2.png")
image1 = image1.resize((140,140), Image.ANTIALIAS)
photo1 = ImageTk.PhotoImage(image1)

image2 = Image.open("raspi.png")
image2 = image2.resize((140,140), Image.ANTIALIAS)
photo2 = ImageTk.PhotoImage(image2)

image3 = Image.open("deviation.png")
image3 = image3.resize((140,150), Image.ANTIALIAS)
photo3 = ImageTk.PhotoImage(image3)

frame= tk.Frame(root, bg="black")
frame.place(relx=0.1, rely=0.1, relwidth=0.8,relheight=0.8)

#image3 = Image.open("pots.png")
#test = ImageTk.PhotoImage(image3)
#label4 = tk.Label(image=test)
#label4.image = test
#label4.place(x=0, y=0)


    
    
button = tk.Button(frame, image=photo5,bg="black",fg="black",border = 5, command= house)
button.place(relx=0.1, rely=0.6, relwidth=0.8, relheight=0.25)


button1 = tk.Button(frame, image=photo1, bg="black",fg="black",border = 5, command= LSTM_temp)
button1.place(relx=0.1, rely=0.3, relwidth=0.25, relheight=0.25)

button2 = tk.Button(frame, image=photo2, bg="black",fg="black",border =5, command=raspi)
button2.place(relx=0.4, rely=0.3, relwidth=0.25, relheight=0.25)

button = tk.Button(frame, image=photo3,bg="black",fg="black",border = 5, command=deviation)
button.place(relx=0.7, rely=0.3, relwidth=0.25, relheight=0.25)



root.mainloop()
```
